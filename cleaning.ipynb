```python
import numpy as np
import pandas as pd
import re
```

```python
data = pd.read_csv('data.csv')
```

```python
data['OS'] = (
    data['OS']
    .str.extract(r'(Windows|Mac|Android|Chrome|Ubuntu|DOS)', expand=False)
    .fillna('Others')
)
```

```python
data['OS'].value_counts()
```

```python
data['ROM_type'] = data['ROM_type'].replace('Hard-Disk', 'HDD')
```

```python
data['ROM_type'].value_counts()
```

```python
data['Ram'] = data['Ram'].str.extract(r'(\d+)').astype(int)
```

```python
data['Ram'].value_counts()
```

```python
data['Processor Brand'] = (
    data['processor']
    .str.extract(r'(Intel|AMD|MediaTek|Apple)', expand=False)
    .fillna('Others')
)
```

```python
data['Processor Brand'].value_counts()
```

```python
data['Processor Family'] = (
    data['processor']
    .str.extract(r'(Core i[3579]|Ryzen [3579]|Pentium|Celeron|Athlon|M1|M2|M3)', expand=False)
    .fillna('Others')
)
```

```python
data['Processor Family'].value_counts()
```

```python
core_map = {
    'Dual': 2,
    'Quad': 4,
    'Hexa': 6,
    'Octa': 8
}

def extract_cores(cpu):
    # Numeric cases
    m = re.search(r'(\d+)\s*Cores?', cpu)
    if m:
        return int(m.group(1))

    # Word cases
    for word, num in core_map.items():
        if word in cpu:
            return num

    return None
```

```python
data['Cores'] = data['CPU'].apply(extract_cores)
```

```python
data['Cores'].fillna(0)
```

```python
data['Threads'] = (
    data['CPU']
    .str.extract(r'(\d+)\s*Threads?', expand=False)
    .fillna(0)
    .astype(float)
)
```

```python
data['Performance_Cores'] = (
    data['CPU']
    .str.extract(r'\((\d+)P', expand=False)
    .fillna(0)
)
```

```python
data['Efficiency_Cores'] = (
    data['CPU']
    .str.extract(r'\+\s*(\d+)E', expand=False)
    .fillna(0)
)
```

```python
data['GPU_Brand'] = (
    data['GPU']
    .str.extract(r'(Intel|NVIDIA|AMD|Apple)', expand=False)
    .fillna('Others')
)
```

```python
data['GPU_Brand'].value_counts()
```

```python
data['PPI'] = ((data['resolution_width']**2 + data['resolution_height']**2)**0.5)/data['display_size']
```

```python
data['GPU'].value_counts().head(20)
```

```python
data['GPU_Series'] = (
    data['GPU']
    .str.extract(r'(RTX|GTX|Integrated|Iris|Radeon)', expand=False)
    .fillna('Others')
)
```

```python
data['GPU_Series'].value_counts()
```

```python
storage_map = {
    '32GB': 32,
    '64GB': 64,
    '128GB': 128,
    '256GB': 256,
    '512GB': 512,
    '1TB': 1024,
    '2TB': 2048
}
```

```python
data['ROM'] = data['ROM'].map(storage_map)
```

```python
data['Gaming'] = data['GPU_Series'].isin(['RTX', 'GTX']).astype(int)
```

```python
data['Gaming'].value_counts()
```

```python
data.to_csv('cleaned_data.csv')
```
